```python
import pandas as pd
import numpy as np
import scipy as sc
import matplotlib.pyplot as plt
```

```python
df1 =  pd.read_csv("out_monomer.csv", header=None, names=["RMSD", "docking_score"])
df1.head(1)
hist1 = df1.hist(bins=5, layout=(1, 3), figsize=(10, 2))
```

```python
df2 =  pd.read_csv("out_dimer.csv", header=None, names=["RMSD", "docking_score"])
df2.head(1)
hist2 = df2.hist(bins=5, layout=(1, 3), figsize=(10, 2))
```

```python
import seaborn as sns
import matplotlib.pyplot as plt
# set a grey background (use sns.set_theme() if seaborn version 0.11.0 or above) 
sns.set(style="whitegrid")

fig, axs = plt.subplots(2, 2, figsize=(7, 7), sharey=True)

sns.histplot(data=df1, x="docking_score", kde=False, color="dodgerblue", ax=axs[0, 0], bins=5)
sns.histplot(data=df1, x="RMSD", kde=False, color="teal", ax=axs[1, 0], bins=5)

sns.histplot(data=df2, x="docking_score", kde=False, color="dodgerblue", ax=axs[0, 1], bins=5)
sns.histplot(data=df2, x="RMSD", kde=False, color="teal", ax=axs[1, 1], bins=5)
```

```python
sns.set(font_scale=1.3)

fig, ((ax1, ax2), (ax3, ax4)) = plt.subplots(nrows=2,ncols=2,figsize=(8,8))

sns.histplot(data=df1, x="docking_score", kde=False, color="dodgerblue", ax=ax1, bins=10)
sns.histplot(data=df2, x="docking_score", kde=False, color="dodgerblue", ax=ax2, bins=10)
ax1.set_ylim([0, 15])
ax2.set_ylim([0, 15])


ax1.set_title("Monomer 6W63 docking")
ax2.set_title("Homo 2-mer 6w63 docking")


sns.histplot(data=df1, x="RMSD", kde=False, color="teal", ax=ax3, bins=10)
sns.histplot(data=df2, x="RMSD", kde=False, color="teal", ax=ax4, bins=10)
ax3.set_ylim([0, 20])
ax4.set_ylim([0, 20])

ax3.set_title("Monomer 6W63 docking")
ax4.set_title("Homo 2-mer 6w63 docking")


fig.tight_layout()
plt.savefig("homodimer_monomer_6W63.png", dpi=300)
```

# mean and standard deviation

```python
## monomer

print("Mean RMSD for monomer docking: {:0.2f} ".format(np.mean(df1["RMSD"])))
print("Standard deviation RMSD for monomer docking: {:0.2f} ".format(np.std(df1["RMSD"])), "\n")


print("Mean docking score for monomer docking: {:0.2f} (kcal/mol)".format(np.mean(df1["docking_score"])))
print("Standard deviation docking score for monomer docking: {:0.2f} (kcal/mol)".format(np.std(df1["docking_score"])))
    
```

```python
## Heterodimer

print("Mean RMSD for monomer docking: {:0.2f} ".format(np.mean(df2["RMSD"])))
print("Standard deviation RMSD for monomer docking: {:0.2f} ".format(np.std(df2["RMSD"])), "\n")


print("Mean docking score for monomer docking: {:0.2f} (kcal/mol)".format(np.mean(df2["docking_score"])))
print("Standard deviation docking score for monomer docking: {:0.2f} (kcal/mol)".format(np.std(df2["docking_score"])))
```

# Two-Sample T-test for comparison

```python
from scipy.stats import ttest_ind
```

## docking score

```python
t_stat, p_value = ttest_ind(df1.docking_score, df2.docking_score)
print("T-statistic value: ", t_stat)  
print("P-Value: ", p_value)
```

## RMSD

```python
t_stat, p_value = ttest_ind(df1.RMSD, df2.RMSD)
print("T-statistic value: ", t_stat)  
print("P-Value: ", p_value)
```
