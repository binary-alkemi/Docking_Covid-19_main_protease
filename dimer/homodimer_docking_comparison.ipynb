```python
import pandas as pd
import numpy as np
import scipy as sc
import matplotlib.pyplot as plt
```

```python
df1 =  pd.read_csv("../monomer/alltrial_data_monomer.dat", header=None, names=["binding_ene", "RMSD_lb", "RMSD_ub"])
df1.head(1)
hist1 = df1.hist(bins=5, layout=(1, 3), figsize=(10, 2))
```

```python
df2 =  pd.read_csv("alltrial_data_dimer.dat", header=None, names=["binding_ene", "RMSD_lb", "RMSD_ub"])
df2.head(1)
hist2 = df2.hist(bins=5, layout=(1, 3), figsize=(10, 2))
```

```python
import seaborn as sns
import matplotlib.pyplot as plt
# set a grey background (use sns.set_theme() if seaborn version 0.11.0 or above) 
sns.set(style="whitegrid")

fig, axs = plt.subplots(3, 2, figsize=(7, 7), sharey=True)

sns.histplot(data=df1, x="binding_ene", kde=False, color="dodgerblue", ax=axs[0, 0], bins=5)
sns.histplot(data=df1, x="RMSD_lb", kde=False, color="teal", ax=axs[1, 0], bins=5)
sns.histplot(data=df1, x="RMSD_ub", kde=False, color="olive", ax=axs[2, 0], bins=5)

sns.histplot(data=df2, x="binding_ene", kde=False, color="dodgerblue", ax=axs[0, 1], bins=5)
sns.histplot(data=df2, x="RMSD_lb", kde=False, color="teal", ax=axs[1, 1], bins=5)
sns.histplot(data=df2, x="RMSD_ub", kde=False, color="olive", ax=axs[2, 1], bins=5)
```

```python
sns.set(font_scale=1.3)

fig, ((ax1, ax2), (ax3, ax4), (ax5, ax6)) = plt.subplots(nrows=3,ncols=2,figsize=(8,10))

sns.histplot(data=df1, x="binding_ene", kde=False, color="dodgerblue", ax=ax1, bins=5)
sns.histplot(data=df2, x="binding_ene", kde=False, color="dodgerblue", ax=ax2, bins=5)
ax1.set_ylim([0, 25])
ax2.set_ylim([0, 25])
ax1.set_title("Monomer 6W63 docking")
ax2.set_title("Homodimer 6w63 docking")

sns.histplot(data=df1, x="RMSD_lb", kde=False, color="teal", ax=ax3, bins=5)
sns.histplot(data=df2, x="RMSD_lb", kde=False, color="teal", ax=ax4, bins=5)
ax3.set_ylim([0, 20])
ax4.set_ylim([0, 20])


sns.histplot(data=df1, x="RMSD_ub", kde=False, color="olive", ax=ax5, bins=5)
sns.histplot(data=df2, x="RMSD_ub", kde=False, color="olive", ax=ax6, bins=5)
ax5.set_ylim([0, 20])
ax6.set_ylim([0, 20])

fig.tight_layout()
plt.savefig("homodimer_monomer_6W63.png", dpi=300)
```
